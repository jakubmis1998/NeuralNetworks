```
# SPLA (Simple Perceptron Learning Algorithm)

import numpy as np
from matplotlib import pyplot as plt
import sys

if __name__ == "__main__":
  # 100 próbek 2D przeusniętych o [3, 1]
  A = np.random.randn(100, 2) + [3, 1]
  # Pierwsza i druga kolumna
  Ax = A[:, 0]
  Ay = A[:, 1]
  # plt.scatter(Ax, Ay)

  # 100 innych próbek 2D
  B = np.random.randn(100, 2) + [-3, 0.5]
  Bx = B[:, 0]
  By = B[:, 1]
  # plt.scatter(Bx, By)

  # Przyklady do uczenia perceptrona
  E = np.zeros((200, 2))
  E[0:100] = A
  E[100:200] = B

  # Odpowiedzi do uczenia
  T = np.zeros((200,))
  T[0:100] = -1
  T[100:200] = 1

  plt.scatter(E[:, 0], E[:, 1], c=T)
  # sys.exit("Error message")

  learning = True
  learning_rate = 0.1
  counter = 0
  k = 100
  N = 2
  # Random weights
  weights = np.random.rand(1, N + 1)
  # weights[0][0] = -THETA
  # Processed examples
  processed = np.full(k * N, False)
  while learning:
    # Random index of examples
    # Dodac losowanie bez powtorzen
    # j = np.random.randint(0, 200)
    for j in range(200):

      O = weights[0][0]
      # Result of the neural operation on E[j]
      for i in range(1, N):
        O += weights[0][i] * E[j, i]
      if O >= 0:
        O = 1
      else:
        O = -1

      ERR = T[j] - O

      if ERR != 0:
        # Count of processed examples
        processed = np.full(k * N, False)
        counter = 0
        # i -> 0, 1, 2  0...N
        for i in range(N):
          weights[0][i] += learning_rate * ERR * E[j, i]
      else:
        # Jesli jeszcze nieprzetworzony
        if not processed[j]:
          counter += 1
          processed[j] = True
      if (j == 199):
        learning = False


# Jaka jest początkowa THETA?
# Skad wziac punkty do narysowania prostej?
```
